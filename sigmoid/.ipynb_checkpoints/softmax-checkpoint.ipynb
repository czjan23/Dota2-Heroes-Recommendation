```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.optimizers import Adagrad
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X= []
Y = []

with open('train.txt','r') as openFile:
    for data in openFile:  
        data = data.split()
        temp = []
        for d in data:
            temp.append(int(d))
        X.append(temp)
with open('label.txt','r') as openFile:
    for data in openFile:  
        Y.append(int(data))
X = np.array(X)
Y = np.array(Y)


feature_train, feature_test, label_train, label_test =train_test_split(X, Y, test_size=0.2)


```

```python
loss_func = ["binary_crossentropy","categorical_crossentropy","poisson","cosine_proximity"]
opti= ["Adagrad"]
activa = ['relu',"linear","tanh","elu","softplus"]


for lo in loss_func[0:1]:
    for op in opti:
        for ac in activa[1:2]:            
            model = Sequential()
            model.add(Dense(500,input_dim=234,activation='relu'))
            model.add(Dense(500,activation='relu'))
            model.add(Dense(500,activation='relu'))
            #model.add(Dense(400,activation='relu'))
            
 
            model.add(Dense(1,activation='sigmoid'))

            epochs = 20
            model.compile(loss=lo, optimizer=op)
            x = model.fit(feature_train, label_train, epochs=epochs, batch_size=800,verbose=1)

            

# count = 0
# for i in range(len(feature_test)):
#     top = list(feature_test[i][0:117])
#     bot = list(feature_test[i][117:234])
#     test = []
#     test.append(list(feature_test[i]))#regular    
#     test.append(bot + top) #flip top and bot
#     test = np.array(test)
    
#     prediction = model.predict(test) 
#     radiant_win = (prediction[0] + 1-prediction[1])/2
    
#     if(radiant_win >= 0.5 and label_test[i] == 1):
#         count += 1
#     elif(radiant_win<0.5 and label_test[i]==0):
#         count += 1   
# print(count/len(label_test))
            
            
            
```

```python
feature_test = []
label_test = []
with open('feature_test.txt','r') as openFile:
    for data in openFile:  
        data = data.split()
        temp = []
        for d in data:
            temp.append(int(d))
        feature_test.append(temp)
with open('label_test.txt','r') as openFile:
    for data in openFile:
        label_test.append(int(data))



feature_test = np.array(feature_test)
label_test = np.array(label_test)


count = 0
for i in range(len(feature_test)):
    top = list(feature_test[i][0:117])
    bot = list(feature_test[i][117:234])
    test = []
    test.append(list(feature_test[i]))#regular    
    test.append(bot + top) #flip top and bot
    test = np.array(test)
    
    prediction = model.predict(test) 
    radiant_win = (prediction[0] + 1-prediction[1])/2
    
    if(radiant_win >= 0.5 and label_test[i] == 1):
        count += 1
    elif(radiant_win<0.5 and label_test[i]==0):
        count += 1   
print(count/len(label_test))
```
